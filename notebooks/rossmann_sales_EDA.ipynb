```python
# Add the src directory to sys.path
import sys
from pathlib import Path
sys.path.append(str(Path.cwd().parent / "scripts"))
```

```python

# Import necessary modules
from eda_pipeline import load_data, clean_data
from visualization import plot_promo_distribution, plot_sales_vs_customers

```

```python
# Load datasets
train_data_path = Path.cwd().parent / "data/rossmann-store-sales/train.csv"
store_data_path = Path.cwd().parent / "data/rossmann-store-sales/store.csv"

```

```python
train_df = load_data(train_data_path)
store_df = load_data(store_data_path)


```

```python
# Merge train and store data
merged_df = train_df.merge(store_df, on="Store")

```

```python
# Clean the data
cleaned_df = clean_data(merged_df)
```

```python
# Plot promotion distribution
plot_promo_distribution(cleaned_df)

```

```python
# Plot sales vs customers
plot_sales_vs_customers(cleaned_df)
```
